# Feature Engineering

Data Preprocessing

```python
#Handelling the missing values

Total= df.isnull().sum()
Percent= (Total/len(df.isnull()))*100
missing_data= pd.concat([Total,Percent],axis=1, keys= ['Total','Percent']).sort_values(by='Percent',ascending=False)
missing_data.head(20)
```

```python
#dropping columns which have more than 15% missing values

df.drop(["PoolQC","MiscFeature","Alley","Fence","FireplaceQu","LotFrontage"],axis=1, inplace=True)
df.isnull().sum().sort_values(ascending=False)
```

*Thus the column number changed from 82 to 76*

```python
#Handling rest all categorical features which have missing values

features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>0 and df[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
## Replace missing value with 'Missing'
def replace_feature(df,features_nan):
    data=df.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

df=replace_feature(df,features_nan)

df[features_nan].isnull().sum()
```

```python
# Now checking for numerical variables containing missing values
numerical_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

# We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
#filling the missing values with the median of the column
df["GarageYrBlt"].fillna(df["GarageYrBlt"].median(),inplace=True)
print(df["GarageYrBlt"].isnull().sum())
```

```python
df["MasVnrArea"].fillna(df["MasVnrArea"].median(), inplace=True)
print(df["MasVnrArea"].isnull().sum())
```
